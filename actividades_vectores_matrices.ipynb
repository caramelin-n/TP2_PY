## Parte 1 — Vectores

Un **vector** es simplemente una lista ordenada de números. Podés pensarlo como una fila de mediciones: temperatura en distintos momentos, coordenadas de un punto, calificaciones de un alumno, etc.

Por ejemplo, el vector $\mathbf{v} = (3, -1, 2)$ se escribe en Python como `v = [3, -1, 2]`.

Operar con dos vectores significa hacer la misma operación aritmética en cada par de posiciones que se corresponden (primero con primero, segundo con segundo, y así).

En todas las actividades de esta sección podés asumir que los vectores de entrada tienen la misma longitud.

### Actividad 3 — Producto punto (producto escalar)

El **producto punto** de dos vectores produce un único número. La idea es simple: multiplicá cada par de elementos en la misma posición y después sumá todos esos productos.

Por ejemplo: `[1, 2, 3] · [4, 5, 6]` → `1×4 + 2×5 + 3×6` → `4 + 10 + 18` → `32`

Formalmente:
$$\mathbf{a} \cdot \mathbf{b} = \sum_{i=0}^{n-1} a_i \cdot b_i$$

**Dato curioso:** Si el producto punto de dos vectores es `0`, significa que son perpendiculares (que forman 90°) entre sí (podés verificarlo con `u` y `v` en las pruebas).

**Otro dato curioso:** Podemos pensar en el producto punto como una forma de medir la "*cercanía*" de dos vectores. Cuando forman 90°, la "cercanía" es 0, mientras que cuando tienden a apuntar en la misma dirección, el producto punto será mayor. Esta idea tiene aplicaciones en un mecanismo fundamental detrás de los LLM (modelos de lenguaje grande) modernos (ChatGPT, Gemini, etc.): el **embedding**.

## Parte 2 — Matrices

Una **matriz** es una tabla de números organizada en filas y columnas. En Python se representa como una lista de listas, donde cada lista interior es una fila.

Por ejemplo, la siguiente tabla de 2 filas y 3 columnas:

$$M = \begin{pmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \end{pmatrix}$$

se escribe en Python como:
```python
M = [
    [1, 2, 3],   # fila 0
    [4, 5, 6]    # fila 1
]
```

Podés pensarlo como una planilla: `M[i][j]` es la celda de la fila `i` y la columna `j`. Una matriz con `m` filas y `n` columnas tiene dimensión $m \times n$.

### Actividad 4 — Suma y resta de matrices

$$C_{ij} = A_{ij} + B_{ij} \quad \text{para cada celda } (i, j)$$

Por ejemplo, la celda `[0][1]` del resultado es `A[0][1] + B[0][1]`.

La **resta** sigue el mismo principio.

**Ejemplo:** con $A$ de $2 \times 3$ y $B$ de $3 \times 2$, el resultado es de $2 \times 2$.

### Actividad 7 — Transpuesta de una matriz

Trasponer una matriz es como **girarla**: las filas se convierten en columnas y las columnas en filas. La primera fila pasa a ser la primera columna, la segunda fila pasa a ser la segunda columna, y así sucesivamente.

Si $A$ tiene dimensión $m \times n$, su transpuesta $A^T$ tiene dimensión $n \times m$.

**Ejemplo:**
$$A = \begin{pmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \end{pmatrix} \quad \Rightarrow \quad A^T = \begin{pmatrix} 1 & 4 \\ 2 & 5 \\ 3 & 6 \end{pmatrix}$$

Formalmente, el elemento que estaba en la fila `i`, columna `j` de $A$, pasa a estar en la fila `j`, columna `i` de $A^T$: $A^T_{ji} = A_{ij}$

**Propiedad para verificar:** Aplicar la transpuesta dos veces debe devolver la matriz original: $(A^T)^T = A$.

````python
def transpuesta(A: list[list[float]]) -> list[list[float]]:
    ...


# --- Pruebas ---
A = [
    [1, 2, 3],
    [4, 5, 6]
]

At = transpuesta(A)
print('A^T:', At)
# esperado: [[1, 4], [2, 5], [3, 6]]

print('(A^T)^T == A:', transpuesta(At) == A)  # esperado: True

# La transpuesta de una matriz cuadrada simétrica es igual a sí misma
S = [
    [1, 2, 3],
    [2, 5, 6],
    [3, 6, 9]
]
print('S^T == S:', transpuesta(S) == S)  # esperado: True
````
